```python
import pandas as pd 
csv_link='https://docs.google.com/spreadsheets/d/e/2PACX-1vTa7adpGv5lAg1gJD6lvUhhXdibgWpimrX8Lu7Ht-_nsKXCVd4gAyNA2krNE6dEtNEtc067FF9jxdR7/pub?gid=1753016774&single=true&output=csv'
```

```python
df= pd.read_csv(csv_link,usecols=['Model','Attack_Recipe','Original_Accuracy_%','Accuracy_Under_Attack_%','Attack_success_rate_%','Average_perturbed_word_%','Avg_num_queries'])
df.tail()
```

# Bert attack BERT 

```python
# BERT attack BERT 
bertattackbert=df[df['Attack_Recipe']=='BERT-Attack'].groupby('Model').mean().sort_values('Accuracy_Under_Attack_%',ascending= False)
bertattackbert.style.highlight_max(subset=['Original_Accuracy_%',
                                        'Accuracy_Under_Attack_%',
                                        'Average_perturbed_word_%',
                                        'Avg_num_queries'],color='lightgreen').highlight_min(subset=['Attack_success_rate_%'],color='yellow')
```

# PWWS

```python
pwws=df[df['Attack_Recipe']=='PWWS'].groupby('Model').mean().sort_values('Accuracy_Under_Attack_%',ascending= False)
pwws.style.highlight_max(subset=['Original_Accuracy_%',
                                        'Accuracy_Under_Attack_%',
                                        'Average_perturbed_word_%',
                                        'Avg_num_queries'],color='lightgreen').highlight_min(subset=['Attack_success_rate_%'],color='yellow')
```

```python
pwws.round(2).to_csv('pwws.csv')
```
